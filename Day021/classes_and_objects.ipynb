```python
class Statistics:
    def __init__(num, data):
        num.data = data

    def count(c_num):
        return len(c_num.data)

    def sum(s_num):
        return sum(s_num.data)

    def min(mi_num):
        return min(mi_num.data)

    def max(mx_num):
        return max(mx_num.data)

    def range(r_num):
        return max(r_num.data) - min(r_num.data)

    def mean(mn_num):
        return sum(mn_num.data) / len(mn_num.data)

    def median(md_num):
        sorted_data = sorted(md_num.data)
        n = len(sorted_data)
        if n % 2 == 0:
            mid1 = sorted_data[n // 2 - 1]
            mid2 = sorted_data[n // 2]
            return (mid1 + mid2) / 2
        else:
            return sorted_data[n // 2]
    def mode(mode_num):
        frequency = {}
        for value in mode_num.data:
            frequency[value] = frequency.get(value, 0) + 1
        max_count = max(frequency.values())
        mode_value = [key for key, count in frequency.items() if count == max_count]
        return {'mode': mode_value[0], 'count': max_count}

    def std(std_num):
        mean = std_num.mean()
        squared_diff = sum((x - mean) ** 2 for x in std_num.data)
        variance = squared_diff / len(std_num.data)
        return variance ** 0.5

    def var(var_num):
        mean = var_num.mean()
        squared_diff = sum((x - mean) ** 2 for x in var_num.data)
        return squared_diff / len(var_num.data)

    def freq_dist(freq_num):
        frequency = {}
        for value in freq_num.data:
            frequency[value] = frequency.get(value, 0) + 1
        return sorted([(count, value) for value, count in frequency.items()], reverse=True)

    def describe(answers):
        result = f"Count: {answers.count()}\n"
        result += f"Sum: {answers.sum()}\n"
        result += f"Min: {answers.min()}\n"
        result += f"Max: {answers.max()}\n"
        result += f"Range: {answers.range()}\n"
        result += f"Mean: {answers.mean()}\n"
        result += f"Median: {answers.median()}\n"
        result += f"Mode: {answers.mode()}\n"
        result += f"Variance: {answers.var()}\n"
        result += f"Standard Deviation: {answers.std()}\n"
        result += f"Frequency Distribution: {answers.freq_dist()}"
        return result
    
scores = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]
data = Statistics(scores)

# Print the statistics
print(data.describe())
```

```python
class PersonAccount:
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = {}
        self.expenses = {}

    def add_income(self, amount, description):
        self.incomes[description] = amount

    def add_expense(self, amount, description):
        self.expenses[description] = amount

    def total_income(self):
        return sum(self.incomes.values())

    def total_expense(self):
        return sum(self.expenses.values())

    def account_balance(self):
        return self.total_income() - self.total_expense()

    def account_info(self):
        info = f"Account Information for {self.firstname} {self.lastname}:\n"
        info += f"Total Income: {self.total_income()}\n"
        info += f"Total Expense: {self.total_expense()}\n"
        info += f"Account Balance: {self.account_balance()}\n"
        info += "\nIncome Details:\n"
        for desc, amount in self.incomes.items():
            info += f"{desc}: {amount}\n"
        info += "\nExpense Details:\n"
        for desc, amount in self.expenses.items():
            info += f"{desc}: {amount}\n"
        return info

# Example usage:
person = PersonAccount("Halima", "Ladan")

# Adding income and expenses
person.add_income(150000, "Salary")
person.add_income(50000, "Remote Work")
person.add_expense(50000, "shop Rent")
person.add_expense(30000, "Utilities")
person.add_expense(20000, "School Fees")

# Displaying account information
print(person.account_info())
```
